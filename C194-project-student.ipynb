# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
#Import libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
#read the csv
df = pd.read_csv("framingham.csv")
ns = df.loc[df["currentSmoker"]==0]
h_ns = ns.groupby("age")["heartRate"].mean().reset_index()

df = pd.read_csv("framingham.csv")
s = df.loc[df["currentSmoker"]==1]
h_s = s.groupby("age")["heartRate"].mean().reset_index()
=
plt.xlabel("Age")
plt.ylabel("Heart Beat")
plt.title("Smoker VS Non-Smoker")

plt.plot(h_ns["age"], h_ns["heartRate"], label="Non-Smokers", linewidth=3)
plt.plot(h_s["age"], h_s["heartRate"], label="Smokers", linewidth=3)
```
